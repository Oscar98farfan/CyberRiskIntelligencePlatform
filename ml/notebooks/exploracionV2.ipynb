#### Unificaicon Severity

```python
print(resultado["SEVERITY_x"].value_counts())
print(resultado["SEVERITY_y"].value_counts())
```

```python
resultado['SEVERITY'] = np.where(
    resultado['SEVERITY_y'].isna() | (resultado['SEVERITY_y'] == 'UNKNOWN'), 
    resultado['SEVERITY_x'], 
    resultado['SEVERITY_y']
)
```

```python
resultado["SEVERITY"].unique()
```

```python
resultado["SEVERITY"] = resultado["SEVERITY"].replace("UNKNOWN", np.nan)
```
