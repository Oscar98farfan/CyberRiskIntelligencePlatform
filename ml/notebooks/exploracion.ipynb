```python
df['CPE_VENDOR'].value_counts().reset_index().rename(
    columns={'index':'vendor','CPE_VENDOR':'total_cves'}
).to_csv('../data/processed/vendors_unicos.csv', index=False)
```

```python
df['CPE_PRODUCT'].value_counts().reset_index().rename(
    columns={'index':'product','CPE_PRODUCT':'total_cves'}
).to_csv('../data/processed/products_unicos.csv', index=False)
```

```python
df['CWE_ID'].value_counts().reset_index().rename(
    columns={'index':'cwe','CWE_ID':'total_cves'}
).to_csv('../data/processed/cwe_unicos.csv', index=False)
```

```python
con_cpe = df[df['CPE_VENDOR'].notna()]
con_cvss = con_cpe[con_cpe['SCORE_BASE'].notna()]
sin_cvss = con_cpe[con_cpe['SCORE_BASE'].isna()]

print(f'Con CPE + CVSS v3.1: {len(con_cvss):,}')
print(f'Con CPE sin CVSS v3.1: {len(sin_cvss):,}')
print(f'HAS_KEV activos: {con_cvss["HAS_KEV"].sum():,}')
```
